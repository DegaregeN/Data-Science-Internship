```python
import pandas as pd
import re

# Initialize an empty DataFrame to store the filtered columns
filtered_df = pd.DataFrame()

# Define the regular expression pattern to match columns starting with 'Q1' and containing '_9A'
pattern = r'^Q1.*_9A.*'

# Loop through each column in the original DataFrame
for column in df.columns:
    # Check if the column name matches the regex pattern
    if re.search(pattern, column):
        # Truncate the column name to 7 characters
        truncated_column_name = column[:9]
        
        # Add the column to the filtered DataFrame with the truncated column name
        filtered_df[truncated_column_name] = df[column]

# Display the filtered DataFrame
filtered_df
```

```python
# Now filter columns in filtered_df that end with '1'
pattern_end_with_1 = r'.*1$'

# Initialize another DataFrame to store columns ending with '1'
df1 = pd.DataFrame()

# Loop through each column in the filtered DataFrame
for column in filtered_df.columns:
    # Check if the column name ends with '1'
    if re.search(pattern_end_with_1, column):
        # Add the column to the final filtered DataFrame
        df1[column] = filtered_df[column]

# Display the final filtered DataFrame
df1
```

```python
# Rename columns with variable names 
df1.columns = [variable_info.get(col, col) for col in df1.columns]

df1
```

```python
# Now filter columns in filtered_df that end with '2'
pattern_end_with_2 = r'.*2$'

# Initialize another DataFrame to store columns ending with '2'
df2 = pd.DataFrame()

# Loop through each column in the filtered DataFrame
for column in filtered_df.columns:
    # Check if the column name ends with '2'
    if re.search(pattern_end_with_2, column):
        # Add the column to the final filtered DataFrame
        df2[column] = filtered_df[column]

# Display the final filtered DataFrame
df2
```

```python
# Rename columns with variable names 
df2.columns = [variable_info.get(col, col) for col in df2.columns]

df2
```

merging of two dataframes

```python
# Merge the two DataFrames row-wise (stack them on top of each other)
dff = pd.concat([df1, df2], axis=0, ignore_index=True)

# Display the merged DataFrame
dff
```

```python
# Define the threshold for null values (90%)
threshold = 0.9

# Drop columns where the proportion of NaN values is >= 90%
df = dff.loc[:, dff.isnull().mean() < threshold]

# Display the cleaned DataFrame
df
```

```python

# Set a random seed for reproducibility
np.random.seed(5954)

# Function to impute missing values based on the probability distribution of existing values for all columns
def impute_with_probability_all_columns(df):
    for column in df.columns:
        # Get the values excluding NaN for the current column
        non_na_values = df[column].dropna()
        
        # Only impute if there are missing values and at least one non-missing value
        if df[column].isna().sum() > 0 and len(non_na_values) > 0:
            # Sample from the non-NaN values with replacement
            random_sample = np.random.choice(non_na_values, size=df[column].isna().sum(), replace=True)
            
            # Fill NaN values with the sampled values
            df.loc[df[column].isna(), column] = random_sample

# Apply the function to the entire DataFrame
impute_with_probability_all_columns(df)

print("\nDataFrame after imputing missing values in all columns:")
df
```

```python
# Calculate the correlation matrix
corr = df.corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
plt.figure(figsize=(22, 12))

# Draw the heatmap with the mask
sns.heatmap(corr, mask=mask, linewidths=0.2, annot=True, cmap="plasma", square=True)

plt.title('Correlation Heatmap', fontsize=20)
plt.show()
```
